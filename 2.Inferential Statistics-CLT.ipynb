### Central Limit Theorem with a BiModal Distribution

```python
a = np.random.normal(100, 5, 1000) # Normal distribution with mean 100 and std 5 and creating 1000 datapoints
```

```python
b = np.random.normal(80, 5, 1000) # Normal distribution with mean 80 and std 5 and creating 1000 datapoints
```

```python
bm_data = np.append(a, b)
np.random.shuffle(bm_data) # shuffling the distribution to avoid biasness
```

```python
sns.displot(bm_data)
plt.xlim(60,120)
```

```python
bm_2 = bm_data.reshape(1000, 2).mean(axis=1)
sns.displot(bm_2)
plt.xlim(60,120)
```

```python
bm_4 = bm_data.reshape(500, 4).mean(axis=1)
sns.displot(bm_4)
plt.xlim(60,120)
```

```python
bm_10 = bm_data.reshape(200, 10).mean(axis=1)
sns.displot(bm_10)
plt.xlim(60,120)
```
